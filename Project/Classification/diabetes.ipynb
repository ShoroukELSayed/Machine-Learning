```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
x = df.iloc[:, 1:22].values # Features: Extracting all rows for all columns except the last one (label) 
y = df.iloc[:, 0].values # Label: Extracting all rows for the last column (label)
```

```python
print(x.shape)
print(y.shape)
```

```python
plt.figure(figsize=(22, 12))
sns.heatmap(df.corr(), annot=True, cmap="RdBu", )
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print("Trainig Data..")
print("The shape of training features: ", x_train.shape)
print("The shape of training labels: ", y_train.shape)
```

```python
print("Testing Data..")
print("The shape of Testing features: ", x_test.shape)
print("The shape of Testing labels: ", y_test.shape)
```

 Training the Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
training_acc = np.ceil(logistic.score(x_train, y_train) * 100)
print("The Training Accuracy = ", training_acc, "%")
```

```python
y_pred = logistic.predict(x_test)
```

```python
testing_acc = np.ceil(logistic.score(x_test, y_test) * 100)
print("The Testing Accuracy = ", testing_acc, "%")
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(testing_acc)
plt.title(all_sample_title, size = 15)
```

```python
name = np.array(list((pd.get_dummies(data=df)).columns))
j=1
for i in range(21):
      plt.figure(figsize=(10,7))
      plt.scatter(x_test[:,i], y_test, color='red', label='Testing Data' )
      plt.scatter(x_test[:,i], y_pred, color='blue', label='Predicted Data')
      plt.legend (fontsize=20)
      plt.title('Logistic Regression')
      plt.xlabel(name[j] , fontsize=20)
      plt.ylabel('Diabetes_binary' , fontsize=20)
      plt.show()
      j=j+1
```
